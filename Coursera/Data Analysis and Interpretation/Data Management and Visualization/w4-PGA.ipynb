```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
dataset = pd.read_csv("https://raw.githubusercontent.com/rmShoeb/MachineLearning/master/Datasets/diabetes_data.csv")
print(dataset.head())
```

```
features_of_interest = ["Age", "Gender", "delayed healing", "Obesity", "class"]
```

```
copy_data = dataset[features_of_interest]
print(copy_data)
```

```
plt.figure(figsize=(20,6))
sns.distplot(copy_data["Age"])
print(copy_data["Age"].describe())
plt.show()
for feature in features_of_interest[1:]:
  print(copy_data[feature].describe())
  plt.figure(figsize=(4,6))
  sns.countplot(x=feature, data=copy_data)
  plt.show()
```

The distribution plot for "Age" seems bi-modal with peaks around 40 and 50. Rest of the features have only two category. So I am quite not sure if I should call their plots unimodal or skewed.

```
sns.catplot(x="Age", y="Obesity", hue="class", data=copy_data, kind="swarm", height=6, aspect=2)
plt.show()
```

This graph pltos "Age" and "Obesity" and shows if the person is diabetic or not.

The graph clearly shows that obesity has nothing to do with diabetes. But the relation between "Age" and diabetes is not quite clear.
